```python
#Imports
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading txt file
with open("train_data_file.txt") as f:
    X,y = np.array([]), np.array([])
    for l in f:
        row = l.split()
        X = np.append(X,int(row[0]))
        y = np.append(y,int(row[1]))
```

```python
#Plotting The Data
plt.scatter(X,y)
plt.title('DATA')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#Our model consists of one hidden layer neural network and this hidden layer contains only 1 neuron
#We will have only 2 learning parameters which are the weight and bias 
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(1, input_shape=[1]))
#I used mean_squared_error for loss function and Adam as an optimizer
#I used learning rate = 0.01
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.01))
model.summary()
```

```python
#The Model starts to train for 50 epochs as the loss will reach approximately zero
history = model.fit(X,y, epochs=60)
```

```python
#Plotting The Loss during the learning of the model
plt.plot(history.history['loss'])
plt.show()
```

```python
#The Red Line represents our model and it shows a great success  
prediction = model.predict(X)
plt.scatter(X, y)
plt.plot(X, prediction, color='r')
plt.show()
```

```python
#The Prediction function so it can be used by the user by entering (x) and the model will predict (y)
def predict(x):
    result= model.predict([x])
    print("The Predction of x = "+str(x)+" , is y = "+str(result[0][0]))
```

```python
#Testing The prediction Function
predict(200)
```
